```python
import pickle

import matplotlib.pyplot as plt
import numpy as np

from src.adversarial_attacks.utils import config
```

```python
train = np.load("../data_blobs/processed/test4", allow_pickle=True)
```

```python
X_train, y_train = train[0], train[1]
```

```python
n_samples = 5

random_img = np.random.choice(len(X_train), n_samples, replace=False)

plt.figure(figsize=(15, 15))

for i in range(n_samples):
    plt.subplot(1, n_samples, i + 1)
    plt.axis("off")
    plt.imshow(X_train[random_img[i]])
    plt.title(config["classes"][np.argmax(y_train[random_img[i]])]);
```
